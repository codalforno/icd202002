Carregando dados do CSV e configurações para salvar

```python
## Nome dos paths onse estão salvos e vamos salvar os dados processados
PATH_DADOS = '../dados/e_sus/' # Nome do path dos arquivos salvos
PATH_DADOS_PROCESSADOS = '../dados/e_sus_processados/' # Nome do path a salvar os arquivos
FILENAME_PROCESSADOS = 'pb_sus_processados.csv' # Nome do arquivo aonde vamos salvar o arquivo

GRAFIA_ESTADO = 'PARAÍBA' # Para a limpeza de dados, usamos a variável para filtrar todos os dados que pertencem a esse estado. Escrever do mesmo jeito que aparece na base

# Arquivos a processar do estado
files = ['dados-pb.csv']
files = [PATH_DADOS + f for f in files]
```

```python
dflist = []

cols = ['ÿid', 'dataNotificacao', 'dataInicioSintomas', 'dataNascimento',
       'sintomas', 'profissionalSaude', 'cbo', 'condicoes', 'estadoTeste',
       'dataTeste', 'tipoTeste', 'resultadoTeste', 'paisOrigem', 'sexo',
       'estado', 'estadoIBGE', 'municipio', 'municipioIBGE', 'origem', 'cnes',
       'estadoNotificacao', 'estadoNotificacaoIBGE', 'municipioNotificacao',
       'municipioNotificacaoIBGE', 'excluido', 'validado', 'idade',
       'dataEncerramento', 'evolucaoCaso', 'classificacaoFinal']


for file in files:
    dfLoad = pd.read_csv(file, sep = ';', encoding = 'latin-1')
    dfLoad = dfLoad[[c for c in dfLoad.columns if c in cols]]
    dflist.append(dfLoad.copy())

df = pd.concat(dflist)
```

### Filtragem de dados para usar somente os registros considerados como casos positivos

```python
dfPositivo = df[(df.estadoTeste == 'Concluído') & (df.resultadoTeste == 'Positivo')]
```

### Limpar df

Os dados são considerados outliers usando o algoritmo Z-Score

```python
dfOK = dfPositivo.copy()
dfOK['idade'] = dfOK['idade'].apply(pd.to_numeric, errors= 'coerce')
dfOK = dfOK[dfOK['idade'].notna()]
dfOK = dfOK[(np.abs(stats.zscore(dfOK[['idade']])) < 3).all(axis=1)]
```

### Normalizando acentos de cidades e estados

```python
# Quando estadoNotificacao não for 'SÃO PAULO', preencher com dados de estado
dfOK['estadoCoalesce'] = np.where(dfOK['estadoNotificacao'] == GRAFIA_ESTADO, dfOK['estadoNotificacao'], dfOK['estado'])
dfOK['municipioCoalesce'] = np.where(dfOK['estadoNotificacao'] == GRAFIA_ESTADO, dfOK['municipioNotificacao'], dfOK['municipio'])

# Normalizando acentos
dfOK['estado_normalizado'] = dfOK['estadoCoalesce'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').str.lower()
dfOK['municipio_normalizado'] = dfOK['municipioCoalesce'].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').str.lower()
```

```python
dfOK['extractDate'] = dfOK['dataNotificacao'].astype(str)
dfOK['extractDate'] = dfOK['extractDate'].str.extract(r'(^\d{4}-\d{2}-\d{2})')
dfOK['extractDate'] = pd.to_datetime(dfOK['extractDate'], format='%Y-%m-%d', errors='coerce')

dfOK['idadeTrunc'] = dfOK['idade'].astype(float).div(10).astype(int)
dfOK['idadeTrunc'] = dfOK['idadeTrunc'].astype(float).mul(10).astype(int)
```

```python
dfOK[['dataNotificacao', 'extractDate', 'idade', 'idadeTrunc']].head(5)
```

## Selecionando colunas e agrupando informação

Após limpar o dataframe, são selecionadas a serem utilizadas dentro da análise e é salvo dentro do CSV. Dados são salvos por estado

```python
dfOK.columns
```

```python
colSelect = ['dataNotificacao', 'sexo', 'idade', 'estado_normalizado', 'municipio_normalizado', 'extractDate', 'idadeTrunc']

dfFinal = dfOK[[col for col in dfOK.columns if col in colSelect]]
```
