```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
%matplotlib inline
```

```python
## Nome dos paths onse estão salvos e vamos salvar os dados processados
PATH_FAIXA_ETARIAS = '../dados/faixa_etaria/' # Nome do path dos arquivos salvos
FILENAME_FAIXA_ETARIAS = 'faixas_etarias_2020.csv'
dfFE = pd.read_csv(PATH_FAIXA_ETARIAS + FILENAME_FAIXA_ETARIAS, encoding='utf-8', delimiter=';')
```

```python
dfFE
```

```python
# colSelFE = [col for col in dfFE.columns if col.startswith('SP')] + ['Grupo Etário']
colSelFE = [col for col in dfFE.columns] + ['Grupo Etário']
colSelFE

dfSelFE = dfFE[colSelFE].set_index('Grupo Etário')
for col in dfSelFE.columns:
    if col != 'Grupo Etário':
        dfSelFE[col] = dfSelFE[col].astype('long')#.str.replace(".", "")#.astype('long')

# dfSelFE = dfSelFE.T
```

```python
dfSelFE
```

```python
listinha = sorted(dfSelFE.columns)
estados = []
for i,k in zip(listinha[0::2], listinha[1::2]):
    estado = i[0:2]
    estados.append(estado)
    print(f'{estado} {i} {k}')
    dfSelFE[estado] = dfSelFE[i] + dfSelFE[k]
```

```python
dfFaixas = dfSelFE[estados].T
dfFaixas = (dfFaixas.groupby((np.arange(len(dfFaixas.columns)) // 2) * 10, axis=1).sum().T
 .reset_index()
 .rename(columns={'index': 'idadeTrunc'})
 .set_index('idadeTrunc'))
```

```python
PATH_TOWRITE = PATH_FAIXA_ETARIAS + 'faixa_etaria_processada.csv' 
dfFaixas.to_csv(PATH_TOWRITE, sep=';', encoding='utf-8')
```

```python
dfSexo = dfSelFE.sum(axis=0).to_frame().T
```

```python
dfList = []
sexos = ['MASCULINO', 'FEMININO']

for sex in sexos:
    colDf = [col for col in dfSexo.columns if sex in col]
    df = dfSexo[colDf]
    colsRename = {}
    for col in colDf:
        colsRename[col] = col[0:2]
    df = df.rename(columns=colsRename)
    df['sexo'] = sex
    dfList.append(df)
```

```python
PATH_TOWRITE = '../dados/sexo/' + 'sexo_processada.csv' 
pd.concat(dfList).set_index('sexo').to_csv(PATH_TOWRITE, sep=';', encoding='utf-8')
```
